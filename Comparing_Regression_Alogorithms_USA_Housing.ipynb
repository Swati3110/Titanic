```python
x_train, x_test, y_train, y_test = train_test_split(X,Y,train_size = 0.7, random_state = 7) 
```

```python
lr=LinearRegression()
dr=DecisionTreeRegressor()
er=ExtraTreeRegressor()
gr=GradientBoostingRegressor()
rr=RandomForestRegressor()

```

```python
def rmse_score(y_actual,y_pred):
    return np.sqrt(np.mean([(pred-actual)**2 for pred,actual in zip(y_pred,y_actual)]))
```

```python
exp_dict={0:(lr,'linearRegression'), 
         1:(dr,'DecisionTreeRegressor'),
         2:(er,'ExtraTreeRegressor'),
         3: (gr,'GradientBoostingRegressor'),
         4: (rr,'RandomForestRegressor')} 
```

```python
res_ls=[]
```

```python
for k in exp_dict.keys():
    model=exp_dict[k][0]
    name=exp_dict[k][1]
    model.fit(x_train, y_train)
    y_pred=model.predict(x_test)
    r_score=model.score(x_test,y_test)
    rmse=rmse_score(y_test,y_pred)
    res_ls.append((name,rmse,r_score))
```

```python
res_ls
```

```python
res_df=pd.DataFrame(res_ls)
```
